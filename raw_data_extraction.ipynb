```python
def extract_text_between_markers(pdf_path, start_marker, end_marker):
    doc = fitz.open(pdf_path)
    extracted_text = ""
    collecting = False

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text = page.get_text("text")

        if start_marker in text:
            collecting = True
            start_pos = text.find(start_marker) + len(start_marker)
            extracted_text += text[start_pos:]

        if end_marker in text:
            end_pos = text.find(end_marker)
            extracted_text += text[:end_pos]
            collecting = False
            break

        if collecting:
            extracted_text += text

    return extracted_text.strip()
```

```python
def extract_tables_from_pages(pdf_path, pages):
    tables = []
    for page in pages:
        tables.extend(camelot.read_pdf(pdf_path, pages=str(page), flavor='stream'))
    return tables
```

```python
def save_tables_to_text(tables, output_path):
    with open(output_path, 'w') as text_file:
        for table in tables:
            df = table.df
            # Iterate over each row in the DataFrame
            for index, row in df.iterrows():
                # Convert the row to the desired text format
                row_text = ', '.join([f"{col}: {row[col]}" for col in df.columns])
                text_file.write(row_text + "\n")
    print(f"Tables have been saved to text file at {output_path}")
```
